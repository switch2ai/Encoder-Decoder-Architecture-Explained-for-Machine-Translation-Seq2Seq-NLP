```python
Source Language : English
Target Language : Hindi
```

```python
Boy eats an apple  --> Ladke ne seb khaya
```

```python
Source Language  --> Encoder --> Context  --> Decoder --> Target Language
                                 Vector
```

```python
## Encoder :
Input : Boy eats an apple

h0 = zero vector
h1 = Boy
h2 = Boy eats
h3 = Boy eats an
h4 = Boy eats an apple

# h4 --> Context Vector
```

```python
## Decoder : <SoS> Ladke ne seb khaya <EoS>
h0 = h4 --> Context Vector
h1 =  --> <SoS>
h2 =  --> Ladke
h3 =  --> ne
h4 =  --> seb
h5 =  --> khaya
h6 =  --> <EoS>
```

```python
## Teacher Forcing :
>> At each time step of decoder, input is output of previous time-step
>> If at some time step model predicted, wrong output --> from training data labels
>> insted of passing wrong output as input to next time step -->
   we'll pass correct output (from data labels) to next time step

   ## Teacher Forcing only applied during : Training
```

```python
## Pros of Encoder-Decoder :
-- Better suited for Seq2Seq
-- End to End Learning

## Cons of Encoder-Decoder :
-- Information might loss because of compressing information in context vector
-- Accuracy might drop for long sentences
```
